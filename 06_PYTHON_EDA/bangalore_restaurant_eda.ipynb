TASK 1: Is customer engagement evenly distributed across Bangalore restaurants, or concentrated among a smaller number of highly popular restaurants?

TASK 2: ARE THERE ANY OUTLIERS?
I suspect outliers from the histogram. Now we want to visually detect extreme values.

```python
sns.boxplot(data=df,x='votes')

plt.title('vote distribution')
plt.xlabel('votes')


plt.show()
```

```python
Q1 = df['votes'].quantile(0.25)
Q3 = df['votes'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
```

TASK 3: Count the potential outliers

```python
vote_outliers = df[df['votes'] > upper_bound]

outlier_count = len(vote_outliers)

outlier_percentage = (outlier_count / len(df)) * 100

print("Vote Outlier Rows:", outlier_count)
print("Outlier Percentage:", round(outlier_percentage, 2), "%")
```

### Votes Distribution and Outlier Analysis

The votes distribution is strongly right-skewed, with the median vote count (39) substantially lower than the mean (260). Using the IQR method, 13.45% of observations were identified as potential statistical outliers.

These observations were retained because high vote counts may represent genuinely popular restaurants rather than data errors. Median-based comparisons and distribution analysis are therefore used alongside mean values to reduce the risk of misleading conclusions.

**Hypothesis testing**

H₀ (Null hypothesis):
There is no significant difference in ratings between online-order and non-online-order restaurants.

H₁ (Alternative hypothesis):
There is a significant difference in ratings between the two groups.

```python
online_yes = df[df['online_order'] == 'Yes']['rating'].dropna()
online_no = df[df['online_order'] == 'No']['rating'].dropna()

print("Yes:", len(online_yes))
print("No:", len(online_no))
```

```python
pip install scipy
```

```python
from scipy.stats import mannwhitneyu
statistic, p_value = mannwhitneyu(
    online_yes,
    online_no,
    alternative='two-sided'
)

print("U Statistic:", statistic)
print("P-value:", p_value)
```

Restaurants offering online ordering tend to have higher ratings, and the Mann–Whitney U test indicates that the difference between the two groups is statistically significant.

```python
n1 = len(online_yes)
n2 = len(online_no)

rank_biserial = (2 * statistic) / (n1 * n2) - 1

print("Rank-Biserial Correlation:", rank_biserial)

print("Online Yes Median:", online_yes.median())
print("Online No Median:", online_no.median())
```

### Online Ordering and Ratings

The Mann–Whitney U test identified a statistically significant difference in rating distributions between restaurants with and without online ordering (p < 0.05). However, the rank-biserial effect size was small (0.099), and median ratings differed by only 0.1 points (3.8 vs 3.7).

This suggests that while online ordering is associated with slightly higher ratings, the practical relationship with customer ratings is limited. Online ordering alone is unlikely to explain restaurant performance.

Price Segment Performance

Analytical question
Which price segment provides the strongest balance between customer satisfaction and engagement?

```python
conditions = [
    df['approx_cost_for_two_cleaned'] <= 500,
    df['approx_cost_for_two_cleaned'].between(501, 1000),
    df['approx_cost_for_two_cleaned'].between(1001, 2000),
    df['approx_cost_for_two_cleaned'] > 2000
]

segments = ['Budget', 'Mid-range', 'Premium', 'Luxury']

df['price_segment'] = np.select(
    conditions,
    segments,
    default='Unknown'
)
```

```python
price_summary = df.groupby('price_segment').agg(
    restaurant_rows=('name', 'count'),
    mean_rating=('rating', 'mean'),
    median_rating=('rating', 'median'),
    mean_votes=('votes', 'mean'),
    median_votes=('votes', 'median')
).round(2)

print(price_summary)
```

```python
sns.barplot(
    data=df,
    x='price_segment',
    y='votes',
    estimator=np.median
)

plt.title('price segment summary')
plt.xlabel('price_segment')
plt.ylabel('votes')

plt.show()
```

Python EDA supports the SQL finding. Premium restaurants show the strongest customer engagement, followed by luxury restaurants, while also maintaining high ratings. This suggests that the premium segment provides the strongest balance between customer satisfaction and engagement.

Location Patterns

SQL said:

Koramangala 5th Block = high restaurant density and total engagement.
Church Street and Lavelle Road = stronger engagement per restaurant.

Analytical question?

Are location averages being inflated by highly popular restaurant outliers?

```python
location_summary = (
    df.groupby('location')
    .agg(
        restaurant_rows=('name', 'count'),
        mean_votes=('votes', 'mean'),
        median_votes=('votes', 'median'),
        median_rating=('rating', 'median')
    )
    .query('restaurant_rows >= 100')
    .sort_values('median_votes', ascending=False)
    .head(10)
    .round(2)
)

print(location_summary) 
```

### Location-Level Engagement

Location analysis revealed that average engagement can be influenced by highly popular restaurants. Koramangala 5th Block showed a substantial gap between mean votes (767) and median votes (183), suggesting that a smaller number of high-engagement restaurants inflate the location average.

In contrast, Lavelle Road and Church Street recorded much higher median vote counts of 500 and 481 respectively. This supports the SQL finding that these locations demonstrate stronger engagement at the typical restaurant level, while Koramangala 5th Block's strength is driven more by market scale and overall restaurant concentration.

Cuisine-Level Patterns.

Analytical question
Are the dominant cuisines also the strongest performers, or do niche cuisines perform better at a smaller scale?

```python
cuisine_summary = (
    df.groupby('cuisines')
    .agg(
        restaurant_rows=('name', 'count'),
        mean_votes=('votes', 'mean'),
        median_votes=('votes', 'median'),
        mean_rating=('rating', 'mean'),
        median_rating=('rating', 'median')
    )
    .query('restaurant_rows >= 100')
    .sort_values('median_votes', ascending=False)
    .head(10)
    .round(2)
)

print(cuisine_summary)
```

Missing rating pattern

Analytical question:

Are missing ratings concentrated among restaurants with low customer engagement?

```python
rating_missing_summary = (
    df.assign(
        rating_status=np.where(
            df['rating'].isna(),
            'Missing Rating',
            'Available Rating'
        )
    )
    .groupby('rating_status')
    .agg(
        restaurant_rows=('name', 'count'),
        median_votes=('votes', 'median'),
        mean_votes=('votes', 'mean')
    )
    .round(2)
)

print(rating_missing_summary)
```

### Missing Rating Pattern

Missing ratings were strongly associated with low customer engagement. Restaurants with missing ratings had a median of 0 votes and an average of only 2.03 votes, compared with a median of 71 votes among rated restaurants.

This suggests that rating missingness is likely related to insufficient customer review activity rather than occurring completely at random. Therefore, missing ratings were not imputed with the overall mean or median during the analysis.
